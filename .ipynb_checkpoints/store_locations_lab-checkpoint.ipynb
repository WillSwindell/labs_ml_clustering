# Lab: using clustering to find best store locations

Imagine the following situation:<br>
You own a pizza chain, and you collected data about pizza deliveries in a certain neighborhood. The data contains a coordinate of each delivery as a pair *(Latitude, Longitude)*. You do not have any stores in this neighborhood, and driving there each time is too expensive (especially with current gas prices). So you decide to open $K$ new stores in this area. The task is, based on the frequent delivery data, determine the best locations for the new stores.

You need to perform the $K$-means clustering of delivery locations, and otput the best location for $K$ new stores. How would you choose the location of the store within each cluster that minimizes the overall distance between the store and each delivery address? __Explain your idea in a separate cell below.__

### My Thoughts

It would seem that when we are dealing with deliveries by vehicle, it would make the most sense to use Manhattan distance, as dense neighborhoods are very likely to be laid out in a grid-like fashion, or at the very least the distance between two points would not be a direct straight line. Therefore, I believe that using clustering which minimizes combined Manhattan distance will lead to the best results when determininbg where to place new stores.

The data is 2-dimensional and it is easy to plot it to see if the locations of new stores make sense.

## Task 1.

Use $K$-means clustering algorithm to find the best locations for new pizza stores for $K$=2, $K$=3 and $K$=4. The answers should be represented as lists of *(Latitide,Longitude)* tuples for each value of $K$.

You can use the custom code from the k-means demo, or you can use the `sklearn.KMeans` module to perform clustering.

```python
def manhattan(v1,v2):
    res=0
    dimensions=min(len(v1),len(v2))

    for i in range(dimensions):
        res+=abs(v1[i]-v2[i])

    return res

import random

# k-means clustering
def kcluster(rows,distance=manhattan,k=4):
    # Print statement to break up multiple calls to kcluster
    print("Clustering for {} clusters".format(k))
   
    # Determine the minimum and maximum values for each point
    ranges=[(min([row[i] for row in rows]),max([row[i] for row in rows]))
    for i in range(len(rows[0]))]

    # Create k randomly placed centroids
    clusters=[[random.random()*(ranges[i][1]-ranges[i][0])+ranges[i][0]
                            for i in range(len(rows[0]))] for j in range(k)]
  
    lastmatches=None
    bestmatches = None

    for t in range(100):
        print ('Iteration %d' % t)
        bestmatches=[[] for i in range(k)]
    
        # Find which centroid is the closest for each row
        for j in range(len(rows)):
            row=rows[j]
            bestmatch=0
            for i in range(k):
                d=distance(clusters[i],row)
                if d<distance(clusters[bestmatch],row): bestmatch=i
            bestmatches[bestmatch].append(j)

        # If the results are the same as last time, this is complete
        if bestmatches==lastmatches: break
        lastmatches=bestmatches
    
        # Move the centroids to the average of the cluster members
        for i in range(k):
            avgs=[0.0]*len(rows[0])
            if len(bestmatches[i])>0:
                for rowid in bestmatches[i]:
                    for m in range(len(rows[rowid])):
                        avgs[m]+=rows[rowid][m]
                for j in range(len(avgs)):
                    avgs[j]/=len(bestmatches[i])
                clusters[i]=avgs
      
    return bestmatches
```

```python
# clustering code for 2,3 and 4 clusters

# 2 clusters
K2 = 2
c_2 = kcluster(data, distance=manhattan, k=K2)

# 3 clusters
K3 = 3
c_3 = kcluster(data, distance=manhattan, k=K3)

# 4 clusters
K4 = 4
c_4 = kcluster(data, distance=manhattan, k=K4)
```

## Task 2
Visualize clusters by plotting each data point and coloring it with a different color corresponding to the cluster to which it belongs. Also plot the locations of new stores for each value of $K$. Some examples of the final visualizations are given below.

First, we can simply plot all the data from the pizza deliveries, adding a small jitter to show repeated coordinates. 

```python
# Get list of coordinates for x and y
def getVals(index, data):
    vals = []
    for i in range(len(data)):
        vals.append(data[i][index])
    return vals

# Jittering function
def rand_jitter(arr):
    jit = .002
    return arr + np.random.randn(len(arr)) * jit

# Get list of x and y coords
x = getVals(0, data)
y = getVals(1, data)

# Create and display scatter plot
f1 = plt.figure()
plt.scatter(rand_jitter(x), rand_jitter(y))
plt.title("Original Data on Pizza Deliveries")
plt.show()

```

Next, we superimpose the locations for each potential k new shops on the graph from above

```python
# Special function that returns lists of x and y coordinates for centroid points

def getCentroids(clusters):
    x = []
    y = []
    for cluster in clusters:
        xs = []
        ys = []
        for index in cluster:
            xs.append(data[index][0])
            ys.append(data[index][1])
        avg_x = sum(xs)/len(xs)
        avg_y = sum(ys)/len(ys)
        x.append(avg_x)
        y.append(avg_y)
            
    return (x, y)

# Special function to return lists of x and y coordinates for each cluster

def getGroups(clusters):
    x = []
    y = []
    for cluster in clusters:
        xs = []
        ys = []
        for index in cluster:
            xs.append(data[index][0])
            ys.append(data[index][1])
        
        x.append(xs)
        y.append(ys)
    return (x, y)
        

# Create the graph for 2 clusters

f2 = plt.figure()
x2, y2 = getGroups(c_2)
c2 = getCentroids(c_2)
plt.scatter(rand_jitter(x2[0]), rand_jitter(y2[0]), s=10, color='b')
plt.scatter(rand_jitter(x2[1]), rand_jitter(y2[1]), s=10, color='g')
plt.scatter(c2[0], c2[1], marker='x', color = 'r')
plt.title("Clusters for K=2. Centroids shown as Red 'X'")
plt.show()

# Create the graph for 3 clusters

f3 = plt.figure()
x3, y3 = getGroups(c_3)
c3 = getCentroids(c_3)
plt.scatter(rand_jitter(x3[0]), rand_jitter(y3[0]), s=10, color='b')
plt.scatter(rand_jitter(x3[1]), rand_jitter(y3[1]), s=10, color='g')
plt.scatter(rand_jitter(x3[2]), rand_jitter(y3[2]), s=10, color='y')
plt.scatter(c3[0], c3[1], marker='x', color = 'r')
plt.title("Clusters for K=3. Centroids shown as Red 'X'")
plt.show()

# Create the graph for 4 clusters

f4 = plt.figure()
x4, y4 = getGroups(c_4)
c4 = getCentroids(c_4)
plt.scatter(rand_jitter(x4[0]), rand_jitter(y4[0]), s=10, color='b')
plt.scatter(rand_jitter(x4[1]), rand_jitter(y4[1]), s=10, color='g')
plt.scatter(rand_jitter(x4[2]), rand_jitter(y4[2]), s=10, color='y')
plt.scatter(rand_jitter(x4[3]), rand_jitter(y4[3]), s=10, color='m')
plt.scatter(c4[0], c4[1], marker='x', color = 'r')
plt.title("Clusters for K=4. Centroids shown as Red 'X'")
plt.show()


```
